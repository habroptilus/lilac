```python
from lilac.validators.lgbm_adversarial_validator import LgbmAdversarialValidator
```

```python
import pandas as pd
train=pd.read_csv("luigi/nothing/train.csv")
only_train_cols = ['pitchType', 'speed', 'ballPositionLabel', 'ballX', 'ballY', 
                   'dir', 'dist', 'battingType', 'isOuts']
train=train.drop(only_train_cols,axis=1)
test=pd.read_csv("luigi/nothing/test.csv")
train
```

```python
validator=LgbmAdversarialValidator(trainer_flag="basic",
                         folds_generator_flag="stratified_group", 
                         man_drop_cols= ["gameID","startDayTime","atbat_id"], 
                         original_target_col="y")
```

```python
output=validator.run(train,test)
```
